```python
import pandas as pd


covid_india_raw = pd.read_csv("/kaggle/input/virus-data/covid_19_india.csv")
icmr_testing_raw = pd.read_csv("/kaggle/input/virus-data/ICMR_Testing_Data.csv")

covid_india_clean = covid_india_raw.rename(columns={
    'State/UT': 'State',
    'Mortality rate': 'Mortality_rate',
    'Recovery rate': 'Recovery_rate'
})

covid_india_clean = covid_india_clean[['Date', 'State', 'Latitude', 'Longitude',
                                       'Confirmed', 'Active', 'Deaths', 'Mortality_rate',
                                       'Cured', 'Recovery_rate']]

icmr_testing_clean = icmr_testing_raw.rename(columns={
    'TotalSamplesTested': 'Total_Samples_Tested',
    'TotalPositiveCases': 'Total_Positive_Cases',
    'PositiveRatio': 'Positive_Ratio',
    'PerDayPositiveCases': 'Perday_Positive',
    'PerDayTests': 'Perday_Tests'
})

covid_india_clean.to_csv("/kaggle/working/covid_india_clean_final.csv", index=False)
icmr_testing_clean.to_csv("/kaggle/working/icmr_testing_clean_final.csv", index=False)
```
